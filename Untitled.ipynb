```python
df["Phishing Status"] = pd.Categorical(df["Phishing Status"])
df.dtypes
```

```python
sample_df = pd.concat([df[df["Phishing Status"] == 0].sample(10000), df[df["Phishing Status"] == 1]])
sample_df.iloc[:,-1].value_counts()
```

```python
X = sample_df.iloc[:, :-1].values
y = sample_df.iloc[:, -1].values
## Splitting to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
sns.set(font_scale=1.4)
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
plt.ylabel("ground truth")
plt.xlabel("prediction")
```

```python
labels = ["True Neg","False Pos","False Neg","True Pos"]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm, annot=labels, fmt="", cmap='Blues')
plt.ylabel("ground truth")
plt.xlabel("prediction")
```

```python
accuracy_score(y_test, y_pred)
# Accuracy: No of correct Prediction/ Total No of Prediction
# Acc: (TP+TN)/(TP+TN+FP+FN)
```

```python
#plot roc curve
#FPR:  FP/FP+TN
#TPR: TP/TP+FN (sensetivity)
metrics.plot_roc_curve(classifier, X_test, y_test) 
```

```python
#logistic regression
score_list = []
for _ in range(1000):
    sample_df = pd.concat([df[df["Phishing Status"] == 0].sample(10000), df[df["Phishing Status"] == 1]])
    sample_df.iloc[:,-1].value_counts()
    X = sample_df.iloc[:, :-1].values
    y = sample_df.iloc[:, -1].values
    # Splitting the dataset into the Training set and Test set
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
    # Feature Scaling
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    # Training the Logistic Regression model on the Training set
    
    classifier = LogisticRegression(random_state = 0)
    classifier.fit(X_train, y_train)
    # Predicting the Test set results
    y_pred = classifier.predict(X_test)
    # Making the Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    score = accuracy_score(y_test, y_pred)
    score_list.append(score)
```

```python
plt.hist(score_list)
plt.title("Accuracy Distribution over 1000 iteration for: logistic regression")
plt.savefig("LogisticRegression.pdf", bbox_inches = "tight")
```

```python
#  Naive Bayes
score_list_NB = []
for _ in range(1000):
    sample_df = pd.concat([df[df["Phishing Status"] == 0].sample(10000), df[df["Phishing Status"] == 1]])
    sample_df.iloc[:,-1].value_counts()
    X = sample_df.iloc[:, :-1].values
    y = sample_df.iloc[:, -1].values
    # Splitting the dataset into the Training set and Test set
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
    # Feature Scaling
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    # Training the Logistic Regression model on the Training set
    classifier_NB = GaussianNB()
    classifier_NB.fit(X_train, y_train)
    # Predicting the Test set results
    y_pred = classifier_NB.predict(X_test)
    # Making the Confusion Matrix
    
    cm = confusion_matrix(y_test, y_pred)
    score_NB = accuracy_score(y_test, y_pred)
    score_list_NB.append(score_NB)
```

```python
plt.hist(score_list_NB)
plt.title("Accuracy Distribution over 1000 iterations for: Naive Bayes")
plt.savefig("NB.pdf", bbox_inches = "tight")
```

```python
#  linear svm
score_list_linear_svm = []
for _ in range(100):
    sample_df = pd.concat([df[df["Phishing Status"] == 0].sample(10000), df[df["Phishing Status"] == 1]])
    sample_df.iloc[:,-1].value_counts()
    X = sample_df.iloc[:, :-1].values
    y = sample_df.iloc[:, -1].values
    # Splitting the dataset into the Training set and Test set
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
    # Feature Scaling
    from sklearn.preprocessing import StandardScaler
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    # Training 
    classifier_svc_lr = SVC(kernel = 'linear', random_state = 0)
    classifier_svc_lr.fit(X_train, y_train)
    # Predicting the Test set results
    y_pred = classifier_svc_lr.predict(X_test)
    # Making the Confusion Matrix

    cm = confusion_matrix(y_test, y_pred)
    score = accuracy_score(y_test, y_pred)
    score_list_linear_svm.append(score)
    
```

```python
plt.hist(score_list_linear_svm)
plt.title("Accuracy Distribution over 100 iteration for: Linear_svm")
plt.savefig("svm_linear.pdf", bbox_inches = "tight")
```

```python
# Gaussian radial basis function (RBF) svm
score_list_rbf_svm = []
for i in range(100):
    print(f'total iteration: {i}')
    
    sample_df = pd.concat([df[df["Phishing Status"] == 0].sample(10000), df[df["Phishing Status"] == 1]])
    sample_df.iloc[:,-1].value_counts()
    X = sample_df.iloc[:, :-1].values
    y = sample_df.iloc[:, -1].values
    # Splitting the dataset into the Training set and Test set

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
    # Feature Scaling
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    # Training the Logistic Regression model on the Training set
    classifier_svc_rbf = SVC(kernel = 'rbf', random_state = 0)
    classifier_svc_rbf.fit(X_train, y_train)
    # Predicting the Test set results
    y_pred = classifier_svc_rbf.predict(X_test)
    # Making the Confusion Matrix
   
    cm = confusion_matrix(y_test, y_pred)
    score = accuracy_score(y_test, y_pred)
    score_list_rbf_svm.append(score)
```

```python
plt.hist(score_list_rbf_svm)
plt.title("Accuracy Distribution over 100 iteration for rbf_svm")
plt.savefig("svm_rbf.pdf", bbox_inches = "tight")
```
